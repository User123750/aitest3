```python
from langchain.agents.middleware import wrap_model_call, ModelRequest, ModelResponse
from langchain.messages import HumanMessage, AIMessage, SystemMessage
```

```python
basic_model = ChatOpenAI(model="gpt-4o-mini",temperature=0)
advanced_model = ChatOpenAI(model="gpt-4o",temperature=0)
```

```python
@wrap_model_call
def dynamic_model_selection(request:ModelRequest, handler)->ModelResponse:
    env = request.runtime.context.get("env","test")
    if env == "prod":
        model = advanced_model
        print("advanced_llm selected")
    else:
        model = basic_model
        print("basic_llm selected")
    return handler(request.override(model=model))
```

```python
agent2 = create_agent(
model=basic_model, 
tools=[],
middleware=[dynamic_model_selection],
debug=True)
```

```python
resp = agent2.invoke(
    input={"messages": [HumanMessage("c'est quoi un agent ai ?")]
    },
    context={"env": "prod"}
)
```

```python
print(resp['messages'][-1].content)
```

```python
from langchain.tools import tool
```

```python
@tool  
def get_weather(city:str) :
  """ 
  Get the weather for a given city.
    """
  print("weather tool  invoked")
  return {
            "city":city,
            "temperature":20,
            "humidity":50,
            "pressure":120


        }


@tool  
def get_employee_info(employee_name:str) :
  """ 
  Get the info about the give, eployee(salary,seniority).
    """
  print("get_enmployee_info tool  invoked")
  return {
            "name":employee_name,
            "salary":40000,
            "seniority":5,
           


        }

```

```python
from langgraph.checkpoint.memory import MemorySaver

memory = MemorySaver()

agent4 = create_agent(
  model="openai:gpt-4o",
  tools=[get_weather, get_employee_info],
  checkpointer=memory,
  system_prompt="answer the user question using only provided tools"
)
```

```python
config={"configurable":{"thread_id":"1"}}
resp=agent4.invoke(input={"messages":[HumanMessage(content="what is the weather in casablanca ?")]},config=config)
print(resp['messages'][-1].content)
```

```python
config={"configurable":{"thread_id":"1"}}
resp=agent4.invoke(input={"messages":[HumanMessage(content="c'est quoi le salaire de hassane ?")]},config=config)
print(resp['messages'][-1].content)
```

```python
from langchain_tavily import TavilySearch
```

```python
tavily=TavilySearch(max_results=10,search_depth="advanced")
```

```python
from langchain.tools import tool
@tool 
def search_web(query: str) :
  """ 
  search for general current web results"""
  print("search_web tool  invoked with {query}")
  results=tavily.invoke({"query":query})
  return results
  
```

```python
from langgraph.checkpoint.memory import MemorySaver

memory = MemorySaver()

agent5 = create_agent(
  model="openai:gpt-4o",
  tools=[search_web],
  checkpointer=memory,
  system_prompt="answer the user question using only provided tools"
)
```

```python
config={"configurable":{"thread_id":"1"}}
resp=agent5.invoke(input={"messages": [HumanMessage(content=" cherche moi  dans le web quel temps fait il à casablanca ?")]}, config=config)
```

```python
print(display(Markdown(resp['messages'][-1].content)))
```
